```python

import pandas as pd 
```

Using pandas data frames to read the data table from the html webpage   
getting the first table only.   
that contain the data that we need is the analsys 

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
list_df = pd.read_html(url)

df = list_df[0]
df[0:10]
```

remove the not assigned data from the data frame   
group this data by the postal code and borough   
and save it in csv File 


```python
df = df[(df.Borough !='Not assigned')]
df['Neighborhood'] = df.groupby(['Postal Code','Borough'])['Neighborhood'].transform(lambda x: ', '.join(x))
df.to_csv('data.csv')
```

remove the duplicates data from the table 

```python
df=df.drop_duplicates()
df.head()
```

reset the indexing of the table after removing the duplicates 

```python
df1= df.reset_index(drop=True, inplace = True)
df.head()
```

rename the data and edit it 

```python
df.Neighborhood[df['Neighborhood']=='Not assigned'] = df.Borough[df['Neighborhood']=='Not assigned']
df.rename(index=str, columns={"Postal Code": "PostalCode"}, inplace = True)
```
